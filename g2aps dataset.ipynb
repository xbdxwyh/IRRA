```python
import logging

import json

import os.path as osp

from prettytable import PrettyTable
```

```python
with open(osp.join(data.dataset_dir,"train.json"),"r") as f:
    data_train = json.load(f)
```

```python
camera = [1]
```

```python
pid_dict = {key:[] for key in range(len(data.id_key))}
```

```python
pid_dict_ptr = {key:0 for key in range(len(data.id_key))}
```

```python
# Set pid_dict
num = 0
for i,item in enumerate(data_train):
    p_id, _, img_path, isuav = item
    p_id = data.id_key[int(p_id)]
    if isuav not in camera:
        pid_dict[int(p_id)].append(i)
        num += 1
```

```python
data_list = []
train_name_list = []
for i,item in enumerate(data_train):
    p_id, _, img_path, isuav = item
    p_id = data.id_key[int(p_id)]
    if isuav in camera and len(pid_dict[int(p_id)]) != 0:
        temp_data = [p_id, i, img_path, data_train[pid_dict[p_id][pid_dict_ptr[p_id]]][2] ,isuav]
        data_list.append(temp_data)
        train_name_list.append(img_path)
        train_name_list.append(data_train[pid_dict[p_id][pid_dict_ptr[p_id]]][2])
        pid_dict_ptr[p_id] = (pid_dict_ptr[p_id]+1) % len(pid_dict[p_id])
```
