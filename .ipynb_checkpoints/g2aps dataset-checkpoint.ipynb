```python
id_list = [int(i[0]) for i in data_train]+[int(i[0]) for i in data_test]
```

```python
id_key = {key:i for i,key in enumerate(set(id_list))}
```

```python
train_pid_list = [int(i[0]) for i in data_train]
test_pid_list = [int(i[0]) for i in data_test]
```

```python
pid_container = set()    
dataset = []
image_pids = []
image_paths = []
caption_pids = []
captions = []
```

```python
image_id = 0
for p_id, img_path, isuav in zip(train_pid_list,train_img_name_list,train_isuav_list):
    if isuav in camera:
        image_pids.append(id_key[p_id])
        pid_container.add(id_key[p_id])
        image_paths.append(img_path)
    pass
```

```python
num_train_pids, num_train_imgs, num_train_captions = \
    len(pid_container), len(image_pids), len(image_pids)
```
